Example notebook for plotting data.

```python
%pylab inline
plt.rc('figure',figsize=(15,4))
plt.rc('font',**{'size':20})
```

```python
from perceptionneuron.load import *
bodyParts = calc_file_body_parts()
```

```python
# Specify file to load.
fname = '/Users/eddie/Dropbox/Documents/Noitom/Axis Neuron/Motion Files/Eddie F Caeli L MarchCaeli.calc'

# If you wish to plot positions, set cols='X'. For velocities, col='V'.
df = load_calc(fname,cols='X')  # Load data into df.
dt = 1/120  # Estimated frame rate.
t = arange(len(df))*dt  # Define time vector.
```

```python
# Specify body part to plot.
bodyPartToPlot = 'Hips'
bodyPartIx = bodyParts.index(bodyPartToPlot)

fig,ax = plt.subplots()
ax.plot(t,df.values[:,bodyPartIx*3:(bodyPartIx+1)*3])
ax.set(xlim=[t[0],t[-1]])
```
